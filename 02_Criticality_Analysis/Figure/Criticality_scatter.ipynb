```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
rat = pd.read_excel('rat_criticality.xlsx', index_col = [0])
lq = pd.read_excel('loc_quotient.xlsx', index_col = [0])
xbase = pd.read_excel('xbase_critical.xlsx', index_col = [0])
```

#### Flatten the arrays

```python
rat_values = list(rat.values.flatten())
lq_values =  list(lq.values.flatten())
xbase_values = list(xbase.values.flatten())
```

```python
# Calculating the co-realtion coefficient
slope, intercept, r_value1, p_value, std_err = linregress(rat_values, xbase_values)
print(r_value1)
```

```python
font = 'Calibri'
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(rat_values, xbase_values, s=4, c='b', marker='o', facecolors='none') 

plt.xlim([0,0.1])
plt.ylim([0,0.025])

plt.yticks(fontsize=10, fontname=font)
plt.xticks(fontsize=10, fontname=font)


plt.text(0.1, 0.9, f' $\\it{{r}}$ : {r_value:.2f}', 
         transform=plt.gca().transAxes, verticalalignment='top', fontsize=10, 
         fontname= font, bbox=dict(facecolor='white', alpha=0))


plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.xlabel('Rationing based criticality ', fontsize=10, fontname= font)
plt.ylabel('Output based criticality', fontsize=10, fontname= font)


inset_axes = plt.gcf().add_axes([0.65, 0.2, 0.2, 0.2])  # Adjust manually as needed
inset_axes.scatter(rat_values, xbase_values, s=4, c='b', marker='o', facecolors='none')

# Set zoomed-in limits
inset_axes.set_xlim([0, 0.01])
inset_axes.set_ylim([0, 0.01])
inset_axes.grid(True, which='both', linestyle='--', linewidth=0.5)

# Adjust ticks and fonts for inset plot
inset_axes.tick_params(axis='both', which='major', labelsize=8)
inset_axes.set_xticks([0,  0.01])
inset_axes.set_yticks([0,  0.01])


#plt.savefig(f'Criticality_scatter.jpeg', dpi=1200, bbox_inches='tight')
```

```python
# Calculating the co-realtion coefficient
slope, intercept, r_value2, p_value, std_err = linregress(rat_values, lq_values)
print(r_value2)
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(rat_values, lq_values, s=4, c='b', marker='o', facecolors='none') 

plt.xlim([0,0.1])
plt.ylim([0,20])

plt.yticks(fontsize=10, fontname=font)
plt.xticks(fontsize=10, fontname=font)


plt.text(0.1, 0.9, f' $\\it{{r}}$ : {r_value:.2f}', 
         transform=plt.gca().transAxes, verticalalignment='top', fontsize=10, 
         fontname= font, bbox=dict(facecolor='white', alpha=0))


plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.xlabel('Rationing based criticality ', fontsize=10, fontname= font)
plt.ylabel('Location Quotient', fontsize=10, fontname= font)


inset_axes = plt.gcf().add_axes([0.65, 0.2, 0.2, 0.2])  # Adjust manually as needed
inset_axes.scatter(rat_values, lq_values, s=4, c='b', marker='o', facecolors='none')

# Set zoomed-in limits
inset_axes.set_xlim([0, 0.025])
inset_axes.set_ylim([0, 2.5])
inset_axes.grid(True, which='both', linestyle='--', linewidth=0.5)

# Adjust ticks and fonts for inset plot
inset_axes.tick_params(axis='both', which='major', labelsize=8)
inset_axes.set_xticks([0,  0.025])
inset_axes.set_yticks([0,  2.5])
```
